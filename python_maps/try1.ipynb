```python
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
from geopandas import GeoDataFrame
import os
import pandas as pd
import numpy as np
# %load_ext watermark
```

```python
path = os.path.join('states_shpfiles', 'states.shp')
states = gpd.read_file(path)
```

```python
states 
```

```python
# remove Alaska and Hawaii (those states are not needed)
states_contig = states[states.STATE_ABBR != 'AK']
states_contig = states_contig[states_contig.STATE_ABBR != 'HI']
```

```python
states_contig.plot()
```

```python
# read in data points at the lat long level
hg_liq_emf = pd.read_excel('data_spdshts/data_boot_cq_remov_Hg.xlsx')
```

```python
hg_liq_emf.columns
```

```python
# remove liq emf that is "NaN"
hg_liq_emf = hg_liq_emf.dropna(axis=0, how='any') # drop any rows where any value is nan
```

```python
hg_liq_emf.head()
```

```python
test = hg_liq_emf[['Longitude', 'Latitude']].apply(lambda row: 
              Point(row["Longitude"], row["Latitude"]), axis=1)
gpd_hg_emf = gpd.GeoDataFrame({"geometry": test, 
            "liq_emf_mg_mwh": hg_liq_emf["liq_emf_mg_mwh"]})
gpd_hg_tot = gpd.GeoDataFrame({"geometry": test, 
            "liq_mg": hg_liq_emf["liq_emf_mg_mwh"]})
```

```python
gpd_hg_emf.head()
```

```python
gpd_hg_tot.head()
```

```python
gpd_hg_tot.shape[0]
```

```python
states_contig.crs = {'init': 'epsg:4326'} # establishes WGS84 coordinate system
gpd_hg_emf.crs = {'init': 'epsg:4326'}
gpd_hg_tot.crs = {'init': 'epsg:4326'}
```

```python
fig, ax = plt.subplots(1)
base = states_contig.plot(ax=ax, color='gray')
#gpd_hg_emf.plot(ax=base, marker="o", 
#        cmap="Blues", markersize=8, 
#        markeredgecolor="black", alpha=1)
#gpd_hg_tot.plot(ax=base, marker="o", 
#        cmap="Blues", markersize = None, 
#        markeredgecolor="black", alpha=1)
gpd_hg_tot.plot(ax=base, marker="o", s = gpd_hg_tot[1])
_ = ax.axis('off')
ax.set_title("Hg median EMF")
```
